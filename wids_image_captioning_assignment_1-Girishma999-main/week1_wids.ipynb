```python
y=dataset.failure
X=dataset.drop(['failure','id','product_code','attribute_1'], axis=1)
#as failure dont depend on id and product code
```

```python

# Select categorical columns 
categorical_cols = [cname for cname in X.columns if
                        X[cname].dtype == "object"]
print('The categorical variables are :', categorical_cols)

# Select numerical columns
numerical_cols = [cname for cname in X.columns if X[cname].dtype in ['int64', 'float64']]
print('The numerical variables are :', numerical_cols)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_valid.shape
```

```python
X_train[numerical_cols].isnull().sum()
```

```python
y_train.isnull().sum()
```

```python
for df1 in [X_train, X_valid]:
    for col in numerical_cols:
        col_median=X_train[col].median()
        df1[col].fillna(col_median, inplace=True) 
```

```python
X_train[numerical_cols].isnull().sum()
```

```python
X_train[categorical_cols].isnull().sum()
```

```python
X_train.isnull().sum()
```

```python
X_valid.isnull().sum()
```

```python
label_encoder = LabelEncoder()
label_cols_train = pd.DataFrame(label_encoder.fit_transform(X_train[categorical_cols]))
label_cols_valid = pd.DataFrame(label_encoder.transform(X_valid[categorical_cols]))
```

```python
# One-hot encoding removed index; put it back
label_cols_train.index = X_train.index
label_cols_valid.index = X_valid.index
```

```python
# Remove categorical columns (will replace with one-hot encoding)
num_X_train = X_train.drop(categorical_cols, axis=1)
num_X_valid = X_valid.drop(categorical_cols, axis=1)

# Add one-hot encoded columns to numerical features
label_X_train = pd.concat([num_X_train, label_cols_train], axis=1)
label_X_valid = pd.concat([num_X_valid, label_cols_valid], axis=1)
label_X_train.head()
```

```python
cols = label_X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

final_X_train = scaler.fit_transform(label_X_train)

final_X_valid = scaler.transform(label_X_valid)
```

```python
final_X_train = pd.DataFrame(final_X_train, columns=[cols])
final_X_valid = pd.DataFrame(final_X_valid, columns=[cols])
```

```python
final_X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(C=0.001,solver='lbfgs')
lr_model.fit(final_X_train, y_train)

```

```python
lr = lr_model.predict(final_X_valid)
```

```python
score = lr_model.score(final_X_valid, y_valid)
print(score)
```

```python
from sklearn.metrics import roc_auc_score
lr_auc = roc_auc_score(y_valid, lr)
print('LogisticRegression: AUROC = %.3f' % (lr_auc))
```
